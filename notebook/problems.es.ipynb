## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import numpy as np
import random
```

```python
# TODO
num_simulations = 1000

dado_1 = np.random.randint(1, 7, num_simulations)  
dado_2 = np.random.randint(1, 7, num_simulations) 

suma = dado_1 + dado_2

condicion = (suma > 7) | (suma % 2 == 0)

eventos_exitosos = np.sum(condicion)

probabilidad_estimada = eventos_exitosos / num_simulations

print(f"La probabilidad estimada es: {probabilidad_estimada}")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)

# Función para realizar el experimento
def realizar_experimento_con_reemplazo():
    bolas_seleccionadas = [ball_box[random.randint(0, 59)] for _ in range(5)]
    blancas = bolas_seleccionadas.count("White")
    rojas = bolas_seleccionadas.count("Red")
    verdes = bolas_seleccionadas.count("Green")
    return blancas, rojas, verdes

# Simulación del experimento 1000 veces
num_experimentos = 1000
resultados = [realizar_experimento_con_reemplazo() for _ in range(num_experimentos)]

# Calcular las probabilidades
tres_blancas_dos_rojas = np.sum([blancas == 3 and rojas == 2 for blancas, rojas, verdes in resultados]) / num_experimentos
mismo_color = np.sum([blancas == 5 or rojas == 5 or verdes == 5 for blancas, rojas, verdes in resultados]) / num_experimentos

tres_blancas_dos_rojas, mismo_color
# TODO
```
